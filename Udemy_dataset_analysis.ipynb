## 7. Show all courses of Graphic Design where the price is below 100 ?

Here we need to work with price column,but here dtype of price is Object we need to convert it into int, but it has some values as free

```python
df.price.unique()
```

```python
# so we need to replace the Free with 0 

df['price'] = df['price'].replace({'Free':0})
```

```python
# converting the Object type to int

df.price = df.price.astype(int)
```

```python
df.subject.unique()
```

```python
graphic_designdf = df.groupby('subject').get_group(('Graphic Design'))
graphic_designdf[:3]
```

```python
graphic_designdf[graphic_designdf.price < 100]
```

# Or we can do it in below way

```python

df[(df.subject == 'Graphic Design') & (df.price<100)]
```
